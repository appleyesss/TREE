```python
data = pd.read_excel(folder_name + file_name)
```

```python
def get_parent_id(x):
    str_x = str(x)
    if len(str_x) == 1:
        return "0"
    else:  # 取到倒数第三个数
        return str_x[:-2]
```

```python
data.columns = ["节点ID", "节点名称"]
```

```python
data = data[~ data['节点ID'].str.contains("T")]
```

```python
data['上级节点ID'] = data['节点ID'].apply(lambda x : get_parent_id(x))
```

```python
id2name_dict = {}
for i in range(data.shape[0]):
    name = data.loc[i, "节点名称"]
    idx = data.loc[i, "节点ID"]
    id2name_dict[str(idx)] = name
```

```python
data['上级节点名称'] = data['上级节点ID'].apply(lambda x : id2name_dict.get(str(x), '支出'))
```

```python
data['节点属性'] = 1
data['是否被停用'] = 0
```

```python
data["节点深度"]=data["节点ID"].apply(lambda x : (len(str(x))+1) // 2)
```
